```python
import pandas as pd

df = pd.read_csv('Tesla Stock Dataset.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
print(df.info())
```

```python
import matplotlib.pyplot as plt

df['Close'].plot(figsize=(12, 6), title="Évolution du prix de clôture Tesla")
plt.show()
```

```python
df['MA50'] = df['Close'].rolling(window=50).mean()
df['MA200'] = df['Close'].rolling(window=200).mean()
```

```python
plt.figure(figsize=(15, 7))
plt.plot(df['Close'], label='Prix de Clôture')
plt.plot(df['MA50'], label='Moyenne 50 jours', color='orange')
plt.plot(df['MA200'], label='Moyenne 200 jours', color='red')
plt.title('Tesla - Prix et Moyennes Mobiles')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df['Prediction'] = df['Close'].shift(-1)
```

```python
X = np.array(df[['Close']])[:-1] 
y = np.array(df['Prediction'])[:-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)

print("Entraînement terminé !")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test[-50:], label='Prix Réel', color='blue')
plt.plot(predictions[-50:], label='Prix Prédit', color='red', linestyle='--')
plt.title('Comparaison : Réel vs Prédiction (50 derniers jours)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))

print(f"Erreur Moyenne Absolue (MAE) : {mae:.2f} $")
print(f"Erreur Quadratique Moyenne (RMSE) : {rmse:.2f} $")
```

```python
score = r2_score(y_test, predictions)
erreur_moyenne = mean_absolute_error(y_test, predictions)

print(f"Précision du modèle (R²) : {score:.4f}")
print(f"Erreur moyenne par prédiction : {erreur_moyenne:.2f} $")
```
